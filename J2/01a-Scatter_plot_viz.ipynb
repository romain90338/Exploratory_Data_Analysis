For this part, we will use the following dataset: house pricing.csv. We will try to understand this dataset a bit better.


1. Import usefull librairies (pandas, seaborn, matplotlib, numpy)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

2. Give the list of variables present in this dataset as well as their nature (are they qualitative, quantitative, discrete etc...) and their type (float, int, str etc...).

```python
df = pd.read_csv("house_pricing.csv")
```

3. When creating a scatterplot, look at how the _LotArea_ column behaves in relation to the _SalePrices_ column.

```python
sns.scatterplot(data = df, x="SalePrice", y="LotArea")
plt.show()
```

4. Refine your visualization by keeping only those houses that have a _LotArea_ of less than 20,000 square feet and a price of less than $500,000.

```python
ax = sns.scatterplot(data = df, x="SalePrice", y="LotArea")
ax.set_xlim(0, 500000)
ax.set_ylim(0, 20000)
plt.show()
```

5. When creating a point cloud, look at the relationship between the _LotFrontage_ and the _LotArea_.

```python
ax = sns.scatterplot(data = df, x="LotFrontage", y="LotArea")
plt.show()
```

6. In the same way, refine your visualization by keeping only those houses that have a _LotFrontage_ of less than 200 square feet and a _LotArea_ of less than 100,000 square feet.


```python
ax = sns.scatterplot(data = df, x="LotFrontage", y="LotArea")
ax.set_xlim(0, 200)
ax.set_ylim(0, 100000)
plt.show()
```
